```python
columns_to_normalize = ['number_preganent', 'glucose_conc', 'blood_pressure', 'skin_thickness',
       'insulin', 'bmi', 'diabetes_pedigree']
```

```python
# for normalizing columns - perform ( x- mean(x)/ stddev(x))
diabetes[columns_to_normalize] = diabetes[columns_to_normalize].apply(lambda x: (x - x.mean()/x.std()))
diabetes.head()
```
