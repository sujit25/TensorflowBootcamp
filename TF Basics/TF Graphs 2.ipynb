```python
import tensorflow as tf
import numpy as np
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

```python
rand_v1 = np.random.uniform(0 ,100 ,(5,100))
rand_v2 = np.random.uniform(0, 100, (5,100))

a = tf.placeholder(dtype=tf.float32, shape=(None, 100))
b = tf.placeholder(dtype=tf.float32, shape=(None, 100))

addop = a + b
mulop = a * b
with tf.Session() as sess:
    add_res, mul_res = sess.run([addop, mulop], feed_dict={a: rand_v1, b: rand_v2})
    print(add_res)
    print(mul_res)
```
